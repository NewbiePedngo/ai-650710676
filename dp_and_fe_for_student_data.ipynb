##คำนวณค่าเฉลี่ยของคอลัมน์ 'อายุ' โดยไม่นับรวมค่า NaN

## แทนที่ค่าที่หายไป (missing values หรือ NaN) ในคอลัมน์ 'อายุ' ด้วยค่าเฉลี่ยของอายุ (mean_age) ที่คำนวณไว้ก่อนหน้านี้

##เติมค่าที่หายไป (missing values หรือ NaN) ในคอลัมน์ 'คณะ' ด้วยค่าที่พบมากที่สุด (mode) ในคอลัมน์นั้น

```python
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
```

```python
df
```

##สร้างฟังก์ชันรับชื่อคณะและแปลงเป็นชื่อมาตรฐาน เพื่อแก้ไขปัญหาชื่อคณะที่ใช้หลายแบบในข้อมูลเดียวกัน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวกรรม', 'วิศวฯ', 'วิศวะ','Engineering']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

```python
df['คณะ_มาตรฐาน']
```

##แสดง ค่าเฉลี่ยค่าเฉลี่ยอายุ จาก df รวมทั้งค่าที่อาจเป็น outliers

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

##หา ควอร์ไทล์ที่ 1 (Q1) และ ควอร์ไทล์ที่ 3 (Q3) ของข้อมูลอายุ

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

```python
Q1, Q3
```

##คำนวณช่วงควอร์ไทล์ (IQR)

```python
IQR = Q3 - Q1
```

```python
IQR
```

##กำหนดขอบเขตล่าง (lower_bound) และบน (upper_bound) เพื่อใช้ตรวจสอบค่าผิดปกติ

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
#lower_bound, upper_bound
print(lower_bound)
print(upper_bound)
```

##กรองค่าผิดปกติเฉพาะแถวที่มีอายุอยู่ในช่วงที่กำหนด

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df_no_outliers
```

##สร้างข้อมูล Feature จาก,ชั่วโมงเรียน_ต่อสัปดาห์,วันที่ลงทะเบียน โดยแปลงวันที่ลงทะเบียนเป็น datetime และสร้าง features จากวันที่ลงทะเบียน อย่างสุดท้ายคือสร้างช่วงเวลาในวัน จากชั่วโมง

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'])
```

```python
df['วันในสัปดาห์'] = df['วันที่ลงทะเบียน'].dt.day_name()
df['เดือน'] = df['วันที่ลงทะเบียน'].dt.month
df['ไตรมาส'] = df['วันที่ลงทะเบียน'].dt.quarter
```

```python
df['ช่วงชั่วโมงเรียน'] = pd.cut(df['ชั่วโมงเรียน_ต่อสัปดาห์'],
                                 bins=[0, 10, 20, 30, 40],
                                 labels=['น้อย', 'ปานกลาง', 'มาก', 'มากมาก'])
```

```python
df
```

## สร้าง BMI จากส่วนสูงและน้ำหนัก และแบ่งค่า BMI ออกเป็นช่วง ๆ ตามเกณฑ์สุขภาพมาตรฐาน

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
```

```python
df
```

```python
df['สุขภาพ'] = pd.cut(df['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```

```python
df
```

##แสดงข้อมูลดิบ และตรวจสอบปัญหาของข้อมูล จำนวนค่าที่หายไป ค่าของ GPA ที่ผิดปกติ

```python
print("=== ข้อมูลดิบ ===")
print(df)

print("\nปัญหาที่พบ")
print(f"- ข้อมูลหาย - {df.isnull().sum().sum()} ค่า")
print(f"- GPA ผิดปกติ - {df[(df['GPA'] < 0) | (df['GPA'] > 4)]['GPA'].tolist()}")
```

##แก้ไขค่าผิดปกติของ GPA และเติมค่า missing ด้วยค่าเฉลี่ยของ GPA ที่ถูกต้อง

```python
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
```
